
1. **What is a random variable in probability theory?**  
   A random variable is a function that assigns numerical values to outcomes in a sample space of a random experiment. It can be **discrete** (finite/countable) or **continuous** (infinite and uncountable).

2. **What are the types of random variables?**  
   - **Discrete Random Variables**: Take a finite or countable number of values (e.g., number of heads in coin flips).  
   - **Continuous Random Variables**: Take an infinite number of values within a range (e.g., height of people).


3. **What is the difference between discrete and continuous distributions?**  
   - **Discrete Distributions** deal with countable outcomes (e.g., Binomial, Poisson).  
   - **Continuous Distributions** deal with outcomes over a continuous range (e.g., Normal, Exponential).

4. **What are probability distribution functions (PDF)?**  
   - For discrete variables, the **Probability Mass Function (PMF)** assigns probabilities to each possible outcome.  
   - For continuous variables, the **Probability Density Function (PDF)** describes the likelihood of values in a range.

5. **How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**  
   - The **CDF** gives the probability that a random variable is **less than or equal to** a given value:
     
     \[ F(x) = P(X \leq x) \]
     
   - The **PDF/PMF** gives the probability of an **exact** value (discrete case) or the likelihood of an interval (continuous case).

6. **What is a discrete uniform distribution?**  
   A distribution where all possible values have **equal probability** (e.g., rolling a fair die). The probability mass function is:
   
   \[ P(X = x) = \frac{1}{n}, \quad x = 1, 2, ..., n \]

7. **What are the key properties of a Bernoulli distribution?**  
   - A random variable that takes two values: success (1) and failure (0).  
   - Has a probability **p** of success and **1 - p** of failure.
   - PMF:
   
     \[ P(X = x) = p^x (1 - p)^{1-x}, \quad x \in \{0,1\} \]

8. **What is the binomial distribution, and how is it used in probability?**  
   - Models the number of successes in **n** independent Bernoulli trials.  
   - Probability formula:  
     
     \[ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} \]
   
   - Used in experiments with fixed trials and success probability.

9. **What is the Poisson distribution and where is it applied?**  
   - Models the number of events in a fixed time/space when events occur independently.  
   - Formula:  
     
     \[ P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} \]
   
   - Used in traffic flow, call centers, etc.

10. **What is a continuous uniform distribution?**  
    - A distribution where all values in a range \([a, b]\) are equally likely.  
    - PDF:
    
      \[ f(x) = \frac{1}{b - a}, \quad a \leq x \leq b \]

11. **What are the characteristics of a normal distribution?**  
    - Bell-shaped, symmetric around the mean \( \mu \).  
    - Defined by mean \( \mu \) and standard deviation \( \sigma \).  
    - PDF:
      
      \[ f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}} \]

    - 68-95-99.7 Rule applies.

12. **What is the standard normal distribution, and why is it important?**  
    - A normal distribution with mean **0** and standard deviation **1**.  
    - Used to standardize data for comparison.

13. **What is the Central Limit Theorem (CLT), and why is it critical in statistics?**  
    - The CLT states that the sampling distribution of the mean approaches normality as the sample size increases, regardless of the population distribution.  
    - Important because it allows for statistical inference using normal approximations.

14. **How does the Central Limit Theorem relate to the normal distribution?**  
    - Even if a population is **not normal**, the sample mean follows a **normal distribution** when the sample size is large.

15. **What is the application of Z statistics in hypothesis testing?**  
    - Used to test hypotheses when population variance is known.  
    - Helps compare sample means with a known population mean.

16. **How do you calculate a Z-score, and what does it represent?**  
    
    \[ Z = \frac{X - \mu}{\sigma} \]
    
    - Measures how far a value is from the mean in **standard deviations**.

17. **What are point estimates and interval estimates in statistics?**  
    - **Point Estimate**: A single value estimate (e.g., sample mean for population mean).  
    - **Interval Estimate**: A range (e.g., confidence interval) that likely contains the true value.

18. **What is the significance of confidence intervals in statistical analysis?**  
    - Provide a range where the true population parameter is expected to lie with a certain probability.

19. **What is the relationship between a Z-score and a confidence interval?**  
    - Z-scores determine the **margin of error** in confidence intervals:
    
      \[ CI = \bar{x} \pm Z \frac{\sigma}{\sqrt{n}} \]

20. **How are Z-scores used to compare different distributions?**  
    - By converting values into **standard normal form**, Z-scores allow direct comparison between different distributions.

21. **What are the assumptions for applying the Central Limit Theorem?**  
    - Sample size should be **large** (\( n \geq 30 \) is a good rule).  
    - Independent random sampling.  
    - Finite variance in the population.

22. **What is the concept of expected value in a probability distribution?**  
    - The **mean** of a random variableâ€™s possible outcomes, weighted by probabilities.  
    - Formula for discrete cases:  
      
      \[ E[X] = \sum x_i P(x_i) \]

23. **How does a probability distribution relate to the expected outcome of a random variable?**  
    - The expected value is the long-run average outcome when an experiment is repeated many times.



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
import sqlite3

# Create a SQLite database
conn = sqlite3.connect("practical_problems.db")
cursor = conn.cursor()

# Table for storing random variables
cursor.execute("""
CREATE TABLE IF NOT EXISTS RandomVariables (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    value REAL
)
""")

# 1. Generate a random variable
random_var = np.random.randint(1, 100)
cursor.execute("INSERT INTO RandomVariables (value) VALUES (?)", (random_var,))
conn.commit()
print("Random Variable:", random_var)

# 2. Discrete uniform distribution PMF
x = np.arange(1, 7)  # Rolling a fair die
pmf = stats.randint.pmf(x, 1, 7)
plt.bar(x, pmf)
plt.title("Discrete Uniform Distribution PMF")
plt.show()

# 3. Bernoulli distribution PMF
def bernoulli_pmf(p, x):
    return p**x * (1 - p)**(1 - x)
print("Bernoulli PMF (p=0.5, x=1):", bernoulli_pmf(0.5, 1))

# 4. Binomial distribution simulation
binomial_data = np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(binomial_data, bins=11, density=True, alpha=0.6, color='g')
plt.title("Binomial Distribution Histogram")
plt.show()

# 5. Poisson distribution visualization
poisson_data = np.random.poisson(lam=5, size=1000)
plt.hist(poisson_data, bins=15, density=True, alpha=0.6, color='b')
plt.title("Poisson Distribution")
plt.show()

# 6. CDF of discrete uniform distribution
cdf = stats.randint.cdf(x, 1, 7)
plt.plot(x, cdf, marker='o')
plt.title("Discrete Uniform Distribution CDF")
plt.show()

# 7. Continuous uniform distribution simulation
uni_data = np.random.uniform(0, 1, 1000)
plt.hist(uni_data, bins=10, density=True, alpha=0.6, color='r')
plt.title("Continuous Uniform Distribution")
plt.show()

# 8. Standard normal distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.title("Standard Normal Distribution")
plt.show()

# 9. Normal distribution simulation
normal_data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(normal_data, bins=30, density=True, alpha=0.6, color='purple')
plt.title("Normal Distribution Histogram")
plt.show()

# 10. Central Limit Theorem demonstration
samples = [np.mean(np.random.exponential(size=30)) for _ in range(1000)]
plt.hist(samples, bins=30, density=True, alpha=0.6, color='brown')
plt.title("Central Limit Theorem Demonstration")
plt.show()

# 15. Simulating multiple samples from normal distribution for CLT
sample_means = [np.mean(np.random.normal(loc=5, scale=2, size=30)) for _ in range(1000)]
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='teal')
plt.title("Multiple Samples from Normal Distribution - CLT Verification")
plt.show()

# 16. Confidence interval calculation
sample_data = np.random.normal(loc=10, scale=5, size=100)
mean = np.mean(sample_data)
std_err = stats.sem(sample_data)
confidence_interval = stats.t.interval(0.95, len(sample_data)-1, loc=mean, scale=std_err)
print("95% Confidence Interval:", confidence_interval)

# 17. CDF of Poisson Distribution
poisson_cdf = stats.poisson.cdf(np.arange(15), mu=5)
plt.plot(np.arange(15), poisson_cdf, marker='o')
plt.title("Poisson Distribution CDF")
plt.show()

# 18. Expected value of continuous uniform distribution
expected_value = np.mean(uni_data)
print("Expected Value of Continuous Uniform Distribution:", expected_value)

# 19. Compare standard deviations of two datasets
data1 = np.random.normal(loc=10, scale=5, size=100)
data2 = np.random.normal(loc=15, scale=10, size=100)
std1, std2 = np.std(data1), np.std(data2)
print("Standard Deviation Comparison:", std1, std2)
plt.hist(data1, bins=30, alpha=0.5, label='Dataset 1')
plt.hist(data2, bins=30, alpha=0.5, label='Dataset 2')
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.show()

# 20. Range and IQR of a normal distribution
range_data = np.max(normal_data) - np.min(normal_data)
iqr_data = stats.iqr(normal_data)
print("Range:", range_data, "IQR:", iqr_data)

# 21. Z-score normalization and visualization
normalized_data = (normal_data - np.mean(normal_data)) / np.std(normal_data)
plt.hist(normalized_data, bins=30, density=True, alpha=0.6, color='cyan')
plt.title("Z-Score Normalization")
plt.show()

# 22. Skewness and Kurtosis
def skewness_kurtosis(data):
    return stats.skew(data), stats.kurtosis(data)
skew, kurt = skewness_kurtosis(normal_data)
print("Skewness:", skew, "Kurtosis:", kurt)

# 23. Hypothesis Testing using Z-statistics
sample_mean = np.mean(sample_data)
pop_mean = 10
pop_std = 5
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(len(sample_data)))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
print("Z-statistic:", z_stat, "P-value:", p_value)

# 24. Calculate Z-score for a given data point
def calculate_z_score(value, data):
    return (value - np.mean(data)) / np.std(data)
z_score_example = calculate_z_score(12, sample_data)
print("Z-Score for value 12:", z_score_example)

# Close database connection
conn.close()
```
